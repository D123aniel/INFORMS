# Data Summary
Order_id: Numerical id for a person's order \
Waybill_id: Numberical order for a waybill. One order_id can have multiple waybill_id if the first is rejected and then put back into the system. \
dt: Date and time of order creation \
courier_id: Numerical id for a courier \
da_id: Area id of a business district \
poi_id: merchant id \
order_push_time: Unix time when order enters the dispatch system for assignment \
grab_lat/lng: Long/Lat of courier when the order is assigned to them \
is_courier_gradded: Boolean, whether waybill is accepted or rejected \
grab,fetch,arrive_time: Sequence of when courier accepts, collects, then delivers the order \

### Intuitive Stuff:
is_weekend: Boolean, weekend or not \
estimate_arrived_time: System promised deliver time \
is_prebook: Pre-booked in the system or not \
sender_lat/lng: Lat/Long of the pickup point (restaurant) \
recipient_lat/lng: Lat/Long of the dropoff (customer home/business/wherever) \
dispatch_time: Waybill is assigned to courier (can be refused) \
estimate_meal_prepare_time: When meal is ready to get \
platform_order_time: Order sent by customer

```python
import math
# Calculate distances using Haversine Formula
def calc_distance(lat1, long1, lat2, long2):
    R = 6356.752 # Polar Radius of the Earth in km (https://nssdc.gsfc.nasa.gov/planetary/factsheet/earthfact.html)
    # Convert to radians
    lat_rad_1 = math.radians(lat1)
    long_rad_1 = math.radians(long1)
    lat_rad_2 = math.radians(lat2)
    long_rad_2 = math.radians(long2)
    
    delta_1 = lat_rad_2 - lat_rad_1
    delta_2 = long_rad_2 - long_rad_1
    
    inside = math.sin(delta_1/2)**2 + math.cos(lat_rad_1) * math.cos(lat_rad_2) * math.sin(delta_2/2)**2
    dist = 2 * R * math.asin(math.sqrt(inside))
    
    return dist
```
