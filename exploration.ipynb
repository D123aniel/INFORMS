```python
# Added a new column 'rejected' to the all_waybill dataframe. If grab_time, fetch_time, and arrive_time are all 0, set to true. Otherwise set to false.
all_waybill['rejected'] = all_waybill.apply(lambda x: x['grab_time'] == 0 and x['fetch_time'] == 0 and x['arrive_time'] == 0, axis=1)
print(all_waybill['rejected'].value_counts())
```
